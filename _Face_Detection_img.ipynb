```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
img=cv2.imread('C:/Users/panindia/Desktop/Udemy/Computer-Vision-with-Python/DATA/Denis_Mukwege.jpg',0)
```

```python

face_cascade=cv2.CascadeClassifier("C:/Users/panindia/Desktop/Udemy/Computer-Vision-with-Python/DATA/haarcascades/haarcascade_frontalface_default.xml")

```

```python
def detect_face(img):
    
    img_copy=img.copy()
    
    face_rects=face_cascade.detectMultiScale(img_copy)
    for (x,y,w,h) in face_rects:
        cv2.rectangle(img_copy,(x,y),(x+w,y+h),(255,255,255),10)
        
    return img_copy
```

```python
w=detect_face(img)
plt.imshow(w,'gray')
```
